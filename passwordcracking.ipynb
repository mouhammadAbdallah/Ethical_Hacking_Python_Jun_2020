md5brute.py

```
#!/usr/bin/python

from termcolor import colored
import hashlib

hpass=input('md5 hash pass: ')
passlist=input('password list: ')

passfile=open(passlist,'r')

for word in passfile.readlines():
        print(colored(word.strip(),'red'))
        enc_word=word.strip().encode('utf-8')
        md5digest=hashlib.md5(enc_word).hexdigest()
        if md5digest==hpass:
                print(colored('pass found: '+word,'green'))
                exit()
```

```
#!/usr/bin/python

import crypt


def crackpass(cryptword):
        salt=cryptword[0:2]
        dic=open('dictionary.txt','r')
        for word in dic.readlines():
                word=word.strip()
                cryptpass=crypt.crypt(word,salt)
                if cryptword==cryptpass:
                        print("found password: "+word)
                        return
        print('pass not found')

passfile=open('passs.txt','r')
for line in passfile.readlines():
        if ":" in line:
                user=line.split(':')[0]
                cryptword=line.split(':')[1].strip(' ').strip('\n')
                print('cracking pass for: '+user)
                crackpass(cryptword)

```
